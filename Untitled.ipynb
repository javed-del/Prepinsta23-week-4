```python
import pandas as pd

# Assuming 'CP_df' is your original DataFrame
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
CP_df = pd.melt(CP_df, id_vars=id_vars, var_name='year', value_name='population')

# Convert 'year' column to numeric (remove non-numeric characters)
CP_df['year'] = pd.to_numeric(CP_df['year'].str.extract('(\d+)', expand=False))

# Display the unpivoted DataFrame
print(CP_df.head())
```
